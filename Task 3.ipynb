```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpb
import seaborn as sbn
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('IRIS.csv')
```

```python
sbn.scatterplot(data=df, x="petal_length", y="petal_width", hue="species");
```

```python
k_rng = range(1,10)
sse=[]

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['petal_length', 'petal_width']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
mpb.xlabel('k_rng')
mpb.ylabel('Sum of Squared errors')
mpb.plot(k_rng,sse)
```

```python
km = KMeans(n_clusters=3,random_state=0,)
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(150)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python
true_labels = df.species
predicted_labels = df.cluster

cm = confusion_matrix(true_labels, predicted_labels)
class_labels = ['Setosa', 'versicolor', 'virginica']

# plotting confusion matrix
mpb.imshow(cm, interpolation='nearest', cmap=mpb.cm.Blues)
mpb.title('Confusion Matrix')
mpb.colorbar()
tick_marks = np.arange(len(class_labels))
mpb.xticks(tick_marks, class_labels)
mpb.yticks(tick_marks, class_labels)

# Filling matrix with values
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        mpb.text(j,i, str(cm[i][j]), ha='center', va='center', color='Black')

mpb.xlabel('Predicted label')
mpb.ylabel('True label')
mpb.show()
```
